```python
def read_kafka_avro(topic: str, kafka_bootstrap: str = "kafka:9092", schema_file: str = None):
    """
    Read a Kafka topic with Avro serialization (Apicurio) and return a DataFrame.
    
    :param topic: Kafka topic name (e.g., 'olist.customers')
    :param kafka_bootstrap: Kafka bootstrap server address
    :param schema_file: path to Avro schema file (.avsc)
    :return: PySpark DataFrame
    """
    kafka_df = spark.read \
        .format("kafka") \
        .option("kafka.bootstrap.servers", kafka_bootstrap) \
        .option("subscribe", topic) \
        .option("startingOffsets", "earliest") \
        .load()

    if schema_file is None:
        schema_file = f"schemas/{topic.split('.')[-1]}.avsc"

    with open(schema_file) as f:
        avro_schema = f.read()

    df = kafka_df.selectExpr("CAST(key AS STRING) as key") \
                .withColumn("data", from_avro("value", avro_schema)) \
                .select("data.*")
    return df
```
